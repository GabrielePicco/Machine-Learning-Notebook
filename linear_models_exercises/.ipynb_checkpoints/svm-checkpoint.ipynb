```python
#find misclassified example
def find_misclassified(y_pred,y):
    misclassified = [[x1,x2] for c, a, x1, x2 in zip(y_pred, y, X[:,0], X[:,1]) if c != a]
    print "Example misclassified: %s" % len(misclassified)
    if(len(misclassified) > 0):
        plt.scatter(misclassified[:,0], misclassified[:,1], c="purple")
```

```python
   def test_svc_classifier(clf):
    # get the separating hyperplane
    w = clf.coef_[0]
    a = -w[0] / w[1]
    xx = np.linspace(-1, 1)
    yy = a * xx - (clf.intercept_[0]) / w[1]

    # plot the parallels to the separating hyperplane that pass through the
    # support vectors (margin away from hyperplane in direction
    # perpendicular to hyperplane). This is sqrt(1+a^2) away vertically in
    # 2-d.
    margin = 1 / np.sqrt(np.sum(clf.coef_ ** 2))
    yy_down = yy - np.sqrt(1 + a ** 2) * margin
    yy_up = yy + np.sqrt(1 + a ** 2) * margin

    # plot the line, the points, and the nearest vectors to the plane
    plt.clf()
    plt.plot(xx, yy, 'k-')
    plt.plot(xx, yy_down, 'k--')
    plt.plot(xx, yy_up, 'k--')

    print "Weights: %s" % clf.coef_
    print "Intercept: %s" % clf.intercept_[0]
    print "Dual Coefficients: %s" % clf.dual_coef_
    support_vector = clf.support_vectors_
    print "\nSupport vectors learnt: %s" % len(support_vector)
    plt.scatter(X[:,0], X[:,1], c=y)
    plt.scatter(support_vector[:,0], support_vector[:,1], c="yellow")

    y_pred = clf.predict(X)
    C = clf.get_params()['C'] 
    print "Margin error: %s" % len(clf.dual_coef_[np.where(np.absolute(clf.dual_coef_) == C)])
    find_misclassified(y_pred,y)
    print "Accuracy: %s" % accuracy_score(y, y_pred)
```

```python
clf = SVC(kernel='linear') # default C = 1
clf.fit(X, y) 
test_svc_classifier(clf)
```

```python
#with C = 1000
clf = SVC(C = 1000, kernel='linear')
clf.fit(X, y) 

test_svc_classifier(clf)
```

```python
#with C = 0.3
clf = SVC(C = 0.3, kernel='linear')
clf.fit(X, y) 

test_svc_classifier(clf)
```

```python
clf = SVC(kernel='rbf')
clf.fit(X, y) 
```

```python
support_vector = clf.support_vectors_
dual_coef = clf.dual_coef_
print "Intercept: %s" % clf.intercept_[0]
print "Dual weights: %s" % clf.dual_coef_
print "\nSuppot vectors learnt: %s" % len(support_vector)
plt.scatter(X[:,0], X[:,1], c=y)
plt.scatter(support_vector[:,0], support_vector[:,1], c="yellow")
y_pred = clf.predict(X)
C = clf.get_params()['C'] 
print "Margin error: %s" % len(clf.dual_coef_[np.where(np.absolute(clf.dual_coef_) == C)])
find_misclassified(y_pred,y)
print "Accuracy: %s" % accuracy_score(y, y_pred)
```
